```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# 2020년 독서량 파일
```

```python
df = pd.read_csv('data/NL_READ_QY_2020.csv')
```

```python
df
```

```python
df = df.rename(columns={'ANALS_TRGET_YEAR' : 'YEAR', 'ANALS_TRGET_MT' : 'MONTH', 'AREA_NM' : 'AREA', 'AGE_FLAG_NM' : 'AGE', 'SEXDSTN_FLAG_NM' : 'GENDER'})
```

```python
df = df.loc[ : , 'MONTH': ]
```

```python
df
```

```python
df['MONTH'].unique()
```

```python
df['AREA'].unique()
```

```python
df['AGE'].unique()
```

```python
# 월별과 연령대별 독서량 바차트로 나타내기
```

```python
# 월별 독서량
MONTH = df.groupby('MONTH')['READ_QY'].mean()
```

```python
MONTH.to_frame().reset_index()
```

```python
MONTH = MONTH.to_frame()
```

```python
MONTH.loc[ MONTH['READ_QY'] == MONTH['READ_QY'].max() ].index.tolist()[0]
```

```python
# 연령대 독서량
AGE = df.groupby('AGE')['READ_QY'].mean()
```

```python
AGE = AGE.to_frame()
```

```python
max_AGE = AGE.loc[ AGE['READ_QY'] == AGE['READ_QY'].max() ].index.tolist()[0]
```

```python
max_AGE
```

```python
# 월별 연령대 독서량 / 연령대별 월 독서량
```

```python
MONTH_AGE = df[['MONTH','AGE', 'READ_QY']]
```

```python
choice_Month = MONTH_AGE.loc[ MONTH_AGE['MONTH'] == 1 ]
```

```python
choice_Month.sort_values('AGE')
```

```python
# 1월의 연령대별 독서량
choice_Month.groupby('AGE')['READ_QY'].mean().to_frame()
```
