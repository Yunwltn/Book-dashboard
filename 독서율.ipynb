```python
# 월별 연령대 독서율 (셀렉트박스로 월 선택하면 해당 월의 연령대 독서율 차트로 표시)
```

```python
Month_AGE = pd.pivot_table(df, index= ['Month','AGE'], aggfunc= np.mean, values=['READ_RT'])
```

```python
choice_Month = Month_AGE.loc[ Month_AGE['Month'] == 1 ]
```

```python
choice_Month = choice_Month.set_index('AGE')
```
